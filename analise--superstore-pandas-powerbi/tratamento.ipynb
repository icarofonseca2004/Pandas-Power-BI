```python
#comeco tratamento e nomeacoes
import pandas as pd
import numpy as np
tabela = pd.read_csv('Sample - Superstore.csv', sep=',', encoding='latin1')
display(tabela)
```

```python
#tratando e analisando

# qtd linhas e colunas
display(tabela.shape)
display(tabela.columns)

#alterando datas para tipo date
tabela['Order Date'] = pd.to_datetime(tabela['Order Date'])
tabela['Ship Date'] = pd.to_datetime(tabela['Ship Date'])

display(tabela.info())

#verificando duplicatas
tabela.duplicated().sum()

#tabela.describe(include= 'object')
tabela.describe()
```

```python
#analises
#lucro total
lucro = tabela['Profit'].sum()
print(lucro)

#faturamento
faturamento = tabela['Sales'].sum()
print(faturamento)

#Quais categorias mais vendem?
cat_popular = tabela[['Category', 'Quantity']].groupby('Category').sum()
cat_popular = cat_popular.sort_values(by = 'Quantity', ascending = False)
display(cat_popular)

#Quais categorias mais lucram?
cat_lucro = tabela[['Category', 'Profit']].groupby('Category').sum()
cat_lucro = cat_lucro.sort_values(by = 'Profit', ascending = False)
cat_lucro["Profit"] = cat_lucro["Profit"].map("R${:,.2f}".format)
display(cat_lucro)

#Quais estados mais faturam?
estado_lucro = tabela[['State', 'Profit']].groupby('State').sum()
estado_lucro = estado_lucro.sort_values(by = 'Profit', ascending= False)
estado_lucro['Profit'] = estado_lucro['Profit'].map("R${:,.2f}".format)
#display(estado_lucro)

#O desconto impacta negativamente o lucro?

descontos = tabela[['Discount', 'Profit']].groupby('Discount').sum()
descontos = descontos.sort_values(by = 'Profit', ascending = False)
descontos['Profit'] = descontos['Profit'].map("R${:,.2f}".format)
display(descontos)

#Como as vendas evoluíram entre 2014 e 2017?

analise_por_ano = tabela[['Order Date', 'Profit']].groupby('Order Date').sum()
analise_por_ano = analise_por_ano.sort_values(by = 'Order Date', ascending = True)
analise_por_ano['Profit'] = analise_por_ano['Profit'].map("R${:,.2f}".format)
display(analise_por_ano)
```
